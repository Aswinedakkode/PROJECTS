```python
from sklearn.utils import resample
```

```python
df.quality.value_counts()
```

```python
one = df[df.quality == 1]
two = df[df.quality == 2]
three = df[df.quality == 3]
```

```python
two.shape
```

```python
one_oversample = resample(one,replace = True,n_samples = len(two),random_state = 8)
```

```python
one_oversample.shape
```

```python
three_oversample = resample(three,replace = True,n_samples = len(two),random_state = 8)
```

```python
three_oversample.shape
```

```python
df_smote = pd.concat([two,one_oversample,three_oversample],axis = 0)
```

```python
df_smote = df_smote.sample(frac = 1,random_state = 66).reset_index(drop = True) #shuffle
```

```python
df_smote.quality.value_counts()
```

```python
df = df_smote.copy()
```

```python
from sklearn.model_selection import train_test_split
```

```python
y = df.loc[:,'quality']
```

```python
X = df.drop(['quality'],axis = 1)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size =0.25)
```

```python
X_train.shape
```

```python
y_train.shape
```

# MODEL

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
rf = RandomForestClassifier(random_state = 86)
```

```python
params = {'n_estimators':[20,40,60,80,100,120],
          'max_features':['sqrt',0.25,0.50,0.75,1],
          'max_depth': [2,3,4,5,6,7,8,9,None]}
```

```python
search = GridSearchCV(rf,params,n_jobs = -1,cv = 10)
search.fit(X_train,y_train)
search.best_params_
```

```python
rf = RandomForestClassifier(n_estimators = 80,max_features = 1,max_depth = None,random_state = 83)
```

```python
model = rf.fit(X_train,y_train)
```

```python
accuracy_score(model.predict(X_train),y_train)
```

```python
y_pred = model.predict(X_test)
```

##### evaluation

```python
acc = accuracy_score(y_pred,y_test)
print('Accuracy :',acc)
```

```python
pd.DataFrame(confusion_matrix(y_test,y_pred,labels = [1,2,3]),index = [1,2,3],columns = [1,2,3])
```

```python
print(classification_report(y_test,y_pred))
```
